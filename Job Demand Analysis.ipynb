```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# reading data
df = pd.read_csv('Data Jobs in India.csv', encoding = 'unicode_escape')
```

```python
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# extract the month name from 'job_posted_date'
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
```

```python
df_pivot = df.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_pivot
```

```python
# sort job_posted_month_name in df_pivot by job_posted_month value
df_pivot = df_pivot.reset_index()
df_pivot['job_posted_month_no'] = pd.to_datetime(df_pivot['job_posted_month'], format='%B').dt.month
df_pivot = df_pivot.sort_values('job_posted_month_no')
df_pivot = df_pivot.set_index('job_posted_month')
df_pivot = df_pivot.drop(columns='job_posted_month_no')

df_pivot
```
